```python
building_df =  pd.read_csv("de_normalized_data.csv")
building_df.head()
```

```python
# df = bulding_data.dropna(how='all')
building_df['ServerDateTime'] = pd.to_datetime(bulding_df['ServerDateTime'])
building_df['Date'] = building_df['ServerDateTime'].dt.strftime('%d/%m/%Y')
# df['Time'] = df['MessageDateTime'].dt.strftime('%H:%M:%S')
building_df['Time'] = building_df['ServerDateTime'].dt.strftime('%H')
building_df = building_df[building_df.CardNumber > 0]
building_df.head()
```

```python
company_df = building_df[['company','CardNumber', 'DoorName', 'Date','Time']]
company_df.head()
company_df = company_df[company_df.company == 'ABC121']
company_df
```

```python
groupd_df = company_df.groupby(['Time','company'])['CardNumber'].count().reset_index()
groupd_df.head()
```

```python
plt.scatter(groupd_df['CardNumber'].to_numpy(),groupd_df['Time'].to_numpy())
```

```python
X = groupd_df["CardNumber"].values.reshape(-1, 1)
print(X.shape)
```

```python
y = groupd_df['Time'].values.reshape(-1,1)
print(y.shape)
```

```python
# Use sklearn's `train_test_split` to split the data into training and testing

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)
```

```python
# Create the model
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
# Fit the model to the training data. 

model.fit(X_train, y_train)
```

```python
# Calculate the mean_squared_error and the r-squared value
# for the testing data

from sklearn.metrics import mean_squared_error, r2_score

# Use our model to make predictions
predicted = model.predict(X_test)
```

```python
predicted
```

```python
# Score the predictions with mse and r2
mse = mean_squared_error(y_test, predicted)
r2 = r2_score(y_test, predicted)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2 ): {r2}")
```
