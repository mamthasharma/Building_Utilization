```python
# df = bulding_data.dropna(how='all')
df['ServerDateTime'] = pd.to_datetime(df['ServerDateTime'])
df['Date'] = df['ServerDateTime'].dt.strftime('%d/%m/%Y')
# df['Time'] = df['MessageDateTime'].dt.strftime('%H:%M:%S')
df['Time'] = df['ServerDateTime'].dt.strftime('%H')
```

```python
data = df
data.head()

data_filter = df[df.CardNumber > 0]
data_filter.head()
```

```python
X = data_filter["CardNumber"].values.reshape(-1, 1)
print(X.shape)
```

```python
y = data_filter["Time"].values.reshape(-1, 1)
print(y.shape)
```

```python
from sklearn.preprocessing import LabelEncoder

# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y)
encoded_y = label_encoder.transform(y)
```

```python
for label, original_class in zip(encoded_y, y):
    print('Original Class: ' + str(original_class))
    print('Encoded Label: ' + str(label))
    print('-' * 12)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
train_scores = []
test_scores = []
for k in range(1, 20, 2):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    train_score = knn.score(X_train, y_train)
    test_score = knn.score(X_test, y_test)
    train_scores.append(train_score)
    test_scores.append(test_score)
    print(f"k: {k}, Train/Test Score: {train_score:.3f}/{test_score:.3f}")
    
    
plt.plot(range(1, 20, 2), train_scores, marker='o')
plt.plot(range(1, 20, 2), test_scores, marker="x")
plt.xlabel("k neighbors")
plt.ylabel("Testing accuracy Score")
plt.show()
```
